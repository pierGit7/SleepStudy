# SVD WITHOUT STANDARDIZATION
WEIRD GRAPH: We  can  correct  for  differences  in  scale  by  standardization.   When  doing  PCA  ondata with attributes of different scales, it can be very important to standardize thedataset.  We standardize a dataset by ensuring each attribute has a mean of zero (asbefore), but also has a variance of one (i.e.  zero mean and unit variance)

```python
from scipy.linalg import svd

# Subtract mean value from data
Y = X - np.ones((N, 1)) * X.mean(axis=0)

# PCA by computing SVD of Y
U, S, V = svd(Y, full_matrices=False)

# Compute variance explained by principal components
rho = (S * S) / (S * S).sum()

threshold = 0.9

# Verify that more than 90% of the variation is explained by the first 3 principal components
cumulative_variance_explained = np.cumsum(rho)
if cumulative_variance_explained[2] >= threshold:
    print("More than 90% of the variation is explained by the first 3 principal components.")
else:
    print("Less than 90% of the variation is explained by the first 3 principal components.")

for i in range(rho.size):
    if cumulative_variance_explained[i] >= 0.95:
        print(f"The first {i + 1} principal components explain more than 95% of the variation.")
        break

# Plot variance explained
plt.figure()
plt.plot(range(1, len(rho) + 1), rho, "x-")
plt.plot(range(1, len(rho) + 1), np.cumsum(rho), "o-")
plt.plot([1, len(rho)], [threshold, threshold], "k--")
plt.title("Variance explained by principal components")
plt.xlabel("Principal component")
plt.ylabel("Variance explained")
plt.legend(["Individual", "Cumulative", "Threshold"])
plt.grid()
plt.show()

pcs = [0, 1, 2]
legendStrs = ["PC" + str(e + 1) for e in pcs]
c = ["r", "g", "b"]
bw = 0.2
r = np.arange(1, M + 1)
for i in pcs:
    plt.bar(r + i * bw, V[:, i], width=bw) 
plt.xticks(r + bw, attributeNames)
plt.xlabel("Attributes")
plt.ylabel("Component coefficients")
plt.legend(legendStrs)
plt.grid()
plt.title("Wine: PCA Component Coefficients")
plt.show()


```

## SVD WITH STANDARDIZATION

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from sklearn.preprocessing import StandardScaler

# Example: Ensure X is correctly defined with your data
# X = ... (Load your dataset here)

# Standardize the data before PCA
# Subtract mean value from data
X_standardized = X - np.ones((N, 1)) * X.mean(axis=0)
X_standardized = X_standardized * (1 / np.std(X_standardized, 0))
print(np.size(X_standardized))

# PCA by computing SVD of the standardized data
U, S, Vh = svd(X_standardized, full_matrices=False)

# Compute variance explained by principal components
rho = (S * S) / (S * S).sum()

threshold = 0.9

# Verify that more than 90% of the variation is explained by the first 3 principal components
cumulative_variance_explained = np.cumsum(rho)
if cumulative_variance_explained[2] >= threshold:
    print("More than 90% of the variation is explained by the first 3 principal components.")
else:
    print("Less than 90% of the variation is explained by the first 3 principal components.")

# Determine how many components explain more than 95% of the variance
for i in range(rho.size):
    if cumulative_variance_explained[i] >= 0.95:
        print(f"The first {i + 1} principal components explain more than 95% of the variation.")
        break

# Plot variance explained
plt.figure()
plt.plot(range(1, len(rho) + 1), rho, "x-", label="Individual")
plt.plot(range(1, len(rho) + 1), cumulative_variance_explained, "o-", label="Cumulative")
plt.plot([1, len(rho)], [threshold, threshold], "k--", label="Threshold")
plt.title("Variance explained by principal components")
plt.xlabel("Principal component")
plt.ylabel("Variance explained")
plt.legend()
plt.grid()
plt.show()

print(Vh[0,:])

V = Vh.T

# Project the centered data onto principal component space
Z = X_standardized @ V

# Indices of the principal components to be plotted
i = 0
j = 1
k = 2

# Plot PCA of the data
f = plt.figure()
plt.title("Wine data: PCA")
# Z = array(Z)
for c in range(1,4):
    # select indices belonging to class c:
    class_mask = y == c
    plt.plot(Z[class_mask, i], Z[class_mask, j], "o", alpha=0.5)
plt.xlabel("PC{0}".format(i + 1))
plt.ylabel("PC{0}".format(j + 1))

# Output result to screen
plt.show()
# Plot PCA of the data
plot_3d = plt.figure()
plt.title("Wine data: PCA")
ax = plot_3d.add_subplot(projection='3d')  # Create a 3D subplot

# Z = array(Z)
for c in range(1, 4):
    # select indices belonging to class c:
    class_mask = y == c
    ax.plot(Z[class_mask, i], Z[class_mask, j], Z[class_mask, k], "o", alpha=0.8)  # Use ax.plot for 3D plotting

# Set the labels for each axis
ax.set_xlabel("PC{0}".format(i + 1))  # Set label for x-axis
ax.set_ylabel("PC{0}".format(j + 1))  # Set label for y-axis
ax.set_zlabel("PC{0}".format(k + 1))  # Set label for z-axis (corrected)

#plt.legend(Z)
plt.show()  # Display the plot





plt.figure()
att_plot = [f"a{i}" for i in range(M)]
print(att_plot)
pcs = [0, 1, 2]
legendStrs = ["PC" + str(e + 1) for e in pcs]
c = ["r", "g", "b"]
bw = 0.2
r = np.arange(1, M + 1)
for i in pcs:
    plt.bar(r + i * bw, V[:, i], width=bw)

plt.xticks(r + bw, att_plot)
plt.xlabel("Attributes")
plt.ylabel("Component coefficients")
plt.legend(legendStrs)
plt.grid()
plt.title("Wine: PCA Component Coefficients")
plt.show()
```
